```python
import pandas as pd
import numpy as np
```

```python
# loading full dataset into a dataframe
data_f = pd.read_csv("C:/Users/rvrma/Documents/3rd semester/Massive Data Mining/project/ml-1m/ratings.csv")
data_f
```

```python
# dividing datframe into the training and testing data
index = data_f.groupby('userId').apply(lambda x: x.sample(frac=0.8, random_state = 0)).index.get_level_values(1)
# creating testing and training files
train = data_f.iloc[index, :].reset_index(drop = True)
test = data_f.drop(index).reset_index(drop = True)
```

```python
# now copying and saving the test and train files as csv files
train.to_csv('train.csv') 
test.to_csv('test.csv')
```
